### Checking duplicate values in data

There isn't duplicate values in our data.

```python
Q1 = np.quantile(df["Chance_of_admit"], 0.25 )
Q3 = np.quantile( df["Chance_of_admit"] , 0.75)
IQR = Q3 - Q1
IQR
```

### Data Normalization -

```python
Y = df["Chance_of_admit"]
df = df.drop( columns = ["Chance_of_admit"])
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df = pd.DataFrame(scaler.fit_transform(df) , columns = df.columns )
df.head()
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split( df , Y , test_size = 0.2 )
```

### Model building - 

```python
from sklearn.linear_model import LinearRegression , Ridge , Lasso
from sklearn.metrics import mean_squared_error , mean_absolute_error
```

```python
model = LinearRegression()
model.fit(x_train , y_train)
```

```python
model.feature_names_in_
```

```python
model.coef_
```

### Model performance evaluation - 

```python
model.intercept_
```

```python
y_pred = model.predict(x_test)
```

```python
# model training r2 score
r2_score_train = model.score( x_train , y_train )
r2_score_train 
```

```python
# r2 score in model testing 
r2_score = model.score(x_test , y_test)
r2_score
```

```python
# MSE of model - 
mean_squared_error( y_test , y_pred )
```

```python
# MAE of model 
mean_absolute_error( y_test , y_pred)
```

```python
# Adjustedd R2 score of model - 
adj_r2 = 1 - ( ( 1 - r2_score) * (len(Y)-1) / (len(Y) - x_train.shape[1] - 1 ))

adj_r2
```

### Test for Homoscedasticity -

```python
sns.scatterplot(x = y_test , y = y_pred)
plt.ylabel("Actual value")
plt.xlabel("Predicted value")
plt.show()
```

```python
# goldfeld-quandt test - 
from statsmodels.compat import lzip
import statsmodels.stats.api as sms

name = ['F statistic', 'p-value']
test = sms.het_goldfeldquandt(y_train, x_train)
lzip(name, test)
```

From the goldfeld-quandt test:

F Statistic comes out to be 1.12 => Implying minimal difference in variance between groups
p-value of 0.219 indicates that this difference is statistically significant at conventional levels of significance (e.g.,0.05).
 Therefore, we accept the null hypothesis of homoscedasticity, and conclude that there is no strong evidence of heteroscedasticity in the data.

### Multicollinearity check by VIF score - 

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
X_t = pd.DataFrame(x_train , columns=x_train.columns)
vif['Features'] = X_t.columns
vif['VIF'] = [variance_inflation_factor(X_t.values, i) for i in range(X_t.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

In our model data we have high multicollinearity features columns except Research. we should drop high vif values columns in our model.
